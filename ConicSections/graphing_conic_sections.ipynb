```python
import numpy as np
import matplotlib.pyplot as plt
```

# Parabolas

```python
a = 1
h = 1
k = -2
n = 100

x = np.linspace(h-5,h+5,n)

y = a*(x-h)**2 + k

plt.plot(x,y)

plt.grid()
plt.axis('square')
plt.show()
```

```python
a = 1
p = 1/(4*a)
h = 1
k = -2
n = 100

x = np.linspace(h-2,h+2,n)

y = a*(x-h)**2 + k

plt.plot(x,y,label='Parabola')

# Vertex
plt.plot(h,k,'ro',label='Vertex')

# Focus
plt.plot(h,k + p,'go',label='Focus')

# Directrix
plt.plot([x[0],x[-1]],[k-p,k-p],color='orange',label='Directrix')

plt.grid()
plt.legend()

plt.show()
```

# Contours & Meshes

```python
X, Y = np.meshgrid(range(0,10),range(0,15))

plt.subplot(1,2,1)
plt.pcolormesh(X,edgecolor='k',linewidth=.1)

plt.subplot(1,2,2)
plt.pcolormesh(Y,edgecolor='k',linewidth=.1)

plt.show()
```

```python
x = np.linspace(-np.pi,2*np.pi,40)
y = np.linspace(0,4*np.pi,72)

X,Y = np.meshgrid(x,y)

Fxy = np.cos(X) + np.sin(Y)

plt.imshow(Fxy)
plt.show()
```

```python
start = -2
stop = 2
n = 100
x = np.linspace(start,stop,n)

X,Y = np.meshgrid(x,x)
s = 2

gaussian2D = np.exp(-(X**2 + Y**2)/s)

plt.imshow(gaussian2D)
plt.axis('off')
plt.show()
```

# Circles

```python
# Circle Equation: (y - k)^2 + (x - h)^2 = r^2 where (h,k) is the center of the circle.
h = 2
k = -3
r = 3

axlim = r + np.max((abs(h),abs(k)))
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

Fxy = (Y-k)**2 + (X - h)**2 - r**2

plt.contour(X,Y,Fxy,0)
plt.plot([-axlim,axlim],[0,0],'k--')
plt.plot([0,0],[-axlim,axlim],'k--')
plt.plot(h,k,'go')

plt.gca().set_aspect('equal')
plt.show()
```

```python
def plot_circles(h,k):

  r = .5
  rmax = 3
  rinc = .2
  index = 0

  color = np.linspace(0,.9,int((rmax-r)/rinc)+1)

  while r < rmax:
    axlim = r + np.max((abs(h),abs(k)))
    x = np.linspace(-axlim,axlim,100)
    X,Y = np.meshgrid(x,x)

    Fxy = (Y-k)**2 + (X - h)**2 - r**2
    
    plt.contour(X,Y,Fxy,0,colors=[color[index],color[index],color[index]],linewidths=2)

    r += rinc
    index += 1

plot_circles(-1.5,0)
plot_circles(1.5,0)
plt.plot([-1.5,1.5],[0,0],'k',linewidth=3)

plt.gca().set_aspect('equal')
plt.axis('off')
plt.show()
```

```python
# alternate solution
h = [-1.5,1.5]
k = 0

x = np.linspace(-5,5,100)
X,Y = np.meshgrid(x,x)

for r in np.linspace(.5,3,20):
  for hi in h:
    Fxy = (Y - k)**2 + (X - hi)**2 - r**2
    plt.contour(X,Y,Fxy,0,colors=[r/3,r/3,r/3])

plt.plot(h,[0,0],'k',linewidth=3)

plt.gca().set_aspect('equal')
plt.axis('off')
plt.show()
```

# Ellipses

```python
# Ellipse Equation: ((y - k)^2 / b^2) + ((x - h)^2 / a^2) - 1 = 0 where (h,k) is the center of the ellipse.

h = 1
k = 2
a = 2
b = 3

axlim = np.max((a,b)) + np.max((abs(h),abs(k)))
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

ellipse = (X-h)**2 / a**2 + (Y-k)**2 / b**2 - 1

plt.contour(X,Y,ellipse,0)

plt.plot(h,k,'go')

plt.gca().set_aspect('equal')
plt.show()
```
