# Random Numbers

```python
# Uniformly distributed random numbers
nums = np.random.rand(1000)

minvalue = 2
maxvalue = 17

nums = nums*(maxvalue - minvalue) + minvalue

plt.plot(nums,'s')
plt.show()

plt.hist(nums)
plt.show()
```

```python
# Normally distributed random numbers
nums = np.random.randn(1000)

plt.plot(nums,'s',alpha=.5)
plt.show()

plt.hist(nums)
plt.show()
```

```python
nums = np.random.randn(1000)

mean = 15
standard_deviation = 4.3

# Modulate/Normalise these values
nums = nums - np.mean(nums)
nums = nums/np.std(nums)

# Stretching distro by multipling by std deviation
nums = nums*standard_deviation + mean

plt.subplot(121)
plt.plot(nums,'s',alpha=.5)

plt.subplot(122)
plt.hist(nums)
plt.show()

print('This distribution has a mean of %s and a standard deviation of %s' % (np.mean(nums),np.std(nums)))
```

```python
deg = 180

rad = deg*np.pi/180 % (2*np.pi)
rad2 = np.deg2rad(deg)

print('The angle %s degrees is %s radians' % (deg,rad))

rad = np.pi
deg = rad*180/np.pi % 360
deg2 = np.rad2deg(rad)

print('The angle %s radians is %s degrees' % (rad,deg))
```

```python
def angle_convert_plot(input, unit):
  if unit == 'radian':
    plt.plot([0,np.cos(input)],[0,np.sin(input)],'r',linewidth=3)
    plt.title('$\\text{Angle of }%s \\text{ rad} $' % input)
  elif unit == 'degree':
    rad = np.deg2rad(input)
    plt.plot([0,np.cos(rad)],[0,np.sin(rad)],'r',linewidth=3)
    plt.title('$\\text{Angle of }%s ^\circ$' % input)
  else:
    print('Invalid unit')
    return
  
  plt.plot([0,1],[0,0],'r',linewidth=2)
  plt.axis('square')
  plt.axis([-1,1,-1,1])
  plt.grid()
  plt.show()

angle_convert_plot(np.pi/4,'radian')

angle_convert_plot(120,'degree')
```

# The Pythagorean Theorem

```python
a = 3
b = 4
c = np.sqrt(a**2 + b**2)

print('The length of the hypotenuse is %s' % c)

plt.plot([0,a],[0,0],'k')
plt.plot([0,0],[0,b],'k')
plt.plot([0,a],[b,0],'k')

plt.plot([0,.2],[.2,.2],'k')
plt.plot([.2,.2],[.2,0],'k')

plt.text(a/3,-.45,'a = %s' % a,fontsize=15)
plt.text(.1,b/3,'b = %s' % b,fontsize=15)
plt.text(a/2+.1,b/2,'c = %s' % c,fontsize=15)

plt.axis('square')
axlim = np.max([a,b]) + .5
plt.axis([-.5,axlim,-.5,axlim])
plt.axis('off')
plt.show()
```

```python
real = 3
imag = 4j

z = np.complex128(3,4)

print('The complex number is %s' % (z))

mag = np.sqrt(np.real(z)**2 + np.imag(z)**2)
print(mag,np.abs(z))

angle = np.arctan2(np.imag(z),np.real(z))
print(angle,np.angle(z))
```

# Euler's Formula

```python
# e^ik = cos(k) + i*sin(k)

k = np.pi/6
m = 2.3

eul = m*np.exp(1j*k)
cis = m*(np.cos(k) + 1j*np.sin(k))

print(eul)
print(cis)
```

```python
mag = np.abs(eul)
ang = np.angle(eul)

print(m,mag)
print(k,ang)

plt.polar([0,ang],[0,mag],'r')

plt.polar(k,m,'bo')

plt.show()
```

```python
def euler_from_cos_sin(cosine,sine):
  
  m = np.sqrt(cosine**2 + sine**2) # pythagorean theorem
  k = np.arctan2(sine,cosine) # inverse tangent

  plt.polar([0,k],[0,m],'b',linewidth=3)

  plt.title('me$^{i\\phi}$, m=%g, $\\phi$=%g' % (m,k))
  plt.thetagrids([0,45,130,200,222.2])
  plt.show()

euler_from_cos_sin(5,4)
```

```python
n = 200
k = np.random.rand(n)*2*np.pi
m = np.random.rand(n)*50

eul = m*np.exp(1j*k)

plt.figure(facecolor='black')

for i in range(0, n):

  r = np.random.rand()
  col = 'green'

  if r < .4:
    col = 'pink'
  elif r >.4 and r < .8:
    col = 'purple'

  plt.polar([0,np.angle(eul[i])],[0,np.abs(eul[i])],color=col)

plt.axis('off')
plt.show()
```
