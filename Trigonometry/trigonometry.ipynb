# Random Numbers

```python
# Uniformly distributed random numbers
nums = np.random.rand(1000)

minvalue = 2
maxvalue = 17

nums = nums*(maxvalue - minvalue) + minvalue

plt.plot(nums,'s')
plt.show()

plt.hist(nums)
plt.show()
```

```python
# Normally distributed random numbers
nums = np.random.randn(1000)

plt.plot(nums,'s',alpha=.5)
plt.show()

plt.hist(nums)
plt.show()
```

```python
nums = np.random.randn(1000)

mean = 15
standard_deviation = 4.3

# Modulate/Normalise these values
nums = nums - np.mean(nums)
nums = nums/np.std(nums)

# Stretching distro by multipling by std deviation
nums = nums*standard_deviation + mean

plt.subplot(121)
plt.plot(nums,'s',alpha=.5)

plt.subplot(122)
plt.hist(nums)
plt.show()

print('This distribution has a mean of %s and a standard deviation of %s' % (np.mean(nums),np.std(nums)))
```

```python
angles = np.random.rand(100)*2*np.pi

for i in angles:
  plt.plot([0,np.cos(i)],[0,np.sin(i)])

plt.axis('square')
plt.axis('off')
plt.show()
```

```python
deg = 180

rad = deg*np.pi/180 % (2*np.pi)
rad2 = np.deg2rad(deg)

print('The angle %s degrees is %s radians' % (deg,rad))

rad = np.pi
deg = rad*180/np.pi % 360
deg2 = np.rad2deg(rad)

print('The angle %s radians is %s degrees' % (rad,deg))
```

```python
def angle_convert_plot(input, unit):
  if unit == 'radian':
    plt.plot([0,np.cos(input)],[0,np.sin(input)],'r',linewidth=3)
    plt.title('$\\text{Angle of }%s \\text{ rad} $' % input)
  elif unit == 'degree':
    rad = np.deg2rad(input)
    plt.plot([0,np.cos(rad)],[0,np.sin(rad)],'r',linewidth=3)
    plt.title('$\\text{Angle of }%s ^\circ$' % input)
  else:
    print('Invalid unit')
    return
  
  plt.plot([0,1],[0,0],'r',linewidth=2)
  plt.axis('square')
  plt.axis([-1,1,-1,1])
  plt.grid()
  plt.show()

angle_convert_plot(np.pi/4,'radian')

angle_convert_plot(120,'degree')
```

# The Pythagorean Theorem

```python
a = 3
b = 4
c = np.sqrt(a**2 + b**2)

print('The length of the hypotenuse is %s' % c)

plt.plot([0,a],[0,0],'k')
plt.plot([0,0],[0,b],'k')
plt.plot([0,a],[b,0],'k')

plt.plot([0,.2],[.2,.2],'k')
plt.plot([.2,.2],[.2,0],'k')

plt.text(a/3,-.45,'a = %s' % a,fontsize=15)
plt.text(.1,b/3,'b = %s' % b,fontsize=15)
plt.text(a/2+.1,b/2,'c = %s' % c,fontsize=15)

plt.axis('square')
axlim = np.max([a,b]) + .5
plt.axis([-.5,axlim,-.5,axlim])
plt.axis('off')
plt.show()
```

```python
real = 3
imag = 4j

z = np.complex128(3,4)

print('The complex number is %s' % (z))

mag = np.sqrt(np.real(z)**2 + np.imag(z)**2)
print(mag,np.abs(z))

angle = np.arctan2(np.imag(z),np.real(z))
print(angle,np.angle(z))
```

# Graphing resolution for sine, cosine, and tangent

```python
x = np.linspace(0,6*np.pi,100)

plt.plot(x,np.sin(x),'r',label='$\\sin(\\theta)$')
plt.plot(x,np.cos(x),'b',label='$\\cos(\\theta)$')

plt.legend()
plt.xlabel('rad.')
plt.ylabel('function value')
plt.show()
```

```python
th = np.linspace(0,4*np.pi,100)

plt.plot(th,np.tan(th),'k',label='$\\tan(\\theta)$')

plt.legend()
plt.xlabel('rad.')
plt.ylabel('function value')
plt.show()
```

```python
ang = np.random.rand()*2*np.pi
tan = np.tan(ang)
sc = np.sin(ang) / np.cos(ang)

print(ang, tan - sc)
```

```python
thetas = np.linspace(0,2*np.pi,100)
y1 = np.sin(thetas + np.cos(thetas))
y2 = np.cos(thetas + np.sin(thetas))

plt.plot(thetas,y1,'r',label='$\\sin(\\theta + \\cos(\\theta))$')
plt.plot(thetas,y2,'b',label='$\\cos(\\theta + \\sin(\\theta))$')

plt.legend()
plt.show()
```

```python
x = np.linspace(0,2*np.pi,100)
theta = 7*np.pi/6
thetas = np.linspace(0,theta,100)

plt.plot(np.cos(x),np.sin(x),'grey')

plt.plot(np.cos(theta),np.sin(theta),'ko')
plt.plot(np.cos(thetas),np.sin(thetas),'k',linewidth=3)

plt.plot([0,0],[0,np.sin(theta)],'r',linewidth=2,label='sin part')
plt.plot([0,np.cos(theta)],[0,0],'g',linewidth=2,label='cos part')

plt.plot([0,np.cos(theta)],[0,np.sin(theta)],'--',color='grey',alpha=.5)
plt.plot([np.cos(theta),np.cos(theta)],[0,np.sin(theta)],'--',color='grey',alpha=.5)
plt.plot([0,np.cos(theta)],[np.sin(theta),np.sin(theta)],'--',color='r',alpha=.5)

plt.plot([-2,2],[0,0],'--',color='grey',alpha=.5)
plt.plot([0,0],[-2,2],'--',color='grey',alpha=.5)

plt.title('$\\theta = \\frac{7\\pi}{6}$')
plt.axis('square')
plt.axis([-1.25,1.25,-1.25,1.25])
plt.legend()
plt.show()
```
