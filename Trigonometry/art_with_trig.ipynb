```python
import numpy as np
import matplotlib.pyplot as plt
```

# Asteroid Radial Curve

```python
t = np.linspace(-6*np.pi, 6*np.pi, 1000)
a = t

x = a*np.cos(t)**3
y = a*np.sin(t)**3

plt.plot(x,y,'m',linewidth=3)

plt.axis('off')
plt.axis('square')
plt.show()
```

```python
n = range(0,9)
t = np.linspace(-6*np.pi, 6*np.pi, 1000)

for i in n:
    a = t**i
    x = a*np.cos(t)**3
    y = a*np.sin(t)**3
    plt.subplot(3,3,i+1)
    plt.plot(x,y,'k')
    np.ravel(plt.axis('off'))
    np.ravel(plt.axis('square'))

plt.show()
```

```python
# Alternative solution
fix,ax = plt.subplots(3,3,figsize=(10,6))
row,col = np.indices((3,3))

for i in range(9):
    a = t**i
    x = a*np.cos(t)**3
    y = a*np.sin(t)**3

    r = row.ravel()[i]
    c = col.ravel()[i]

    ax[r,c].plot(x,y,'k')
    ax[r,c].axis('off')
    ax[r,c].axis('square')

plt.show()
```

# Rose Curves

```python
k = 4.2
t = np.linspace(0,8*np.pi,1000)
x = np.cos(k*t)*np.cos(t)
y = np.cos(k*t)*np.sin(t)

plt.plot(x,y,'r',linewidth=2)

plt.axis('off')
plt.axis('square')
plt.axis('equal')
plt.show()
```

```python
fix,ax = plt.subplots(3,3,figsize=(10,6))
row,col = np.indices((3,3))

k = 0
for i in range(9):
    t = np.linspace(0,8*np.pi,1000)
    x = np.cos(k*t)*np.cos(t)
    y = np.cos(k*t)*np.sin(t)

    r = row.ravel()[i]
    c = col.ravel()[i]

    ax[r,c].plot(x,y,'k')
    ax[r,c].axis('off')
    ax[r,c].axis('square')
    ax[r,c].set_title('k = {}'.format(k))

    k += 0.5

plt.show()
```
