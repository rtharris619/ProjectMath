```python
import numpy as np
import matplotlib.pyplot as plt
```

# Asteroid Radial Curve

```python
t = np.linspace(-6*np.pi, 6*np.pi, 1000)
a = t

x = a*np.cos(t)**3
y = a*np.sin(t)**3

plt.plot(x,y,'m',linewidth=3)

plt.axis('off')
plt.axis('square')
plt.show()
```

```python
n = range(0,9)
t = np.linspace(-6*np.pi, 6*np.pi, 1000)

for i in n:
    a = t**i
    x = a*np.cos(t)**3
    y = a*np.sin(t)**3
    plt.subplot(3,3,i+1)
    plt.plot(x,y,'k')
    np.ravel(plt.axis('off'))
    np.ravel(plt.axis('square'))

plt.show()
```

```python
# Alternative solution
fix,ax = plt.subplots(3,3,figsize=(10,6))
row,col = np.indices((3,3))

for i in range(9):
    a = t**i
    x = a*np.cos(t)**3
    y = a*np.sin(t)**3

    r = row.ravel()[i]
    c = col.ravel()[i]

    ax[r,c].plot(x,y,'k')
    ax[r,c].axis('off')
    ax[r,c].axis('square')

plt.show()
```

# Rose Curves

```python
k = 4.2
t = np.linspace(0,8*np.pi,1000)
x = np.cos(k*t)*np.cos(t)
y = np.cos(k*t)*np.sin(t)

plt.plot(x,y,'r',linewidth=2)

plt.axis('off')
plt.axis('square')
plt.axis('equal')
plt.show()
```

```python
fix,ax = plt.subplots(3,3,figsize=(10,6))
row,col = np.indices((3,3))

k = 0
for i in range(9):
    t = np.linspace(0,8*np.pi,1000)
    x = np.cos(k*t)*np.cos(t)
    y = np.cos(k*t)*np.sin(t)

    r = row.ravel()[i]
    c = col.ravel()[i]

    ax[r,c].plot(x,y,'k')
    ax[r,c].axis('off')
    ax[r,c].axis('square')
    ax[r,c].set_title('k = {}'.format(k))

    k += 0.5

plt.show()
```

# Squircle

```python
# x**4 + y**4 = a**4
# y = (a**4 - x**4)**1/4

a = 1
x = np.linspace(-1,1,2001)
y1 = (a**4 - x**4)**(1/4)
y2 = -y1

plt.plot(x,y1,'k')
plt.plot(x,y2,'k')
plt.axis('square')
plt.axis('off')
plt.show()
```

# Logarithmic Spirals

```python
t = np.linspace(0,10*np.pi,1234)
k = -3
x = np.cos(t) * np.exp(t/k)
y = np.sin(t) * np.exp(t/k)

plt.plot(x,y,'r')
plt.axis('square')
plt.show()
```

```python
t = np.linspace(0,10*np.pi,1234)
N = 150
ks = np.linspace(-6,-2,N)

for i in range(N):
    k = ks[i]
    x = np.cos(t) * np.exp(t/k)
    y = np.sin(t) * np.exp(t/k)
    plt.plot(x,y,color=[i/N,0,i/N])
    plt.axis('square')
    plt.axis('off')

plt.show()
```

# Logistic Map

```python
# x[n+1] = r*x[n]*(1-x[n])

rs = np.linspace(1,4,1000)
N = 500
x = .5 + np.zeros(N)
endcap = np.arange(round(N*.9),N)

for ri in range(len(rs)):
  r = rs[ri]
  for n in range(N-1):
    # x[n+1] = r*x[n]*(1-x[n])
    x[n+1] = r*x[n]*(1-x[n]) + .2*np.sin(2*np.pi*x[n])**2
  
  u = np.unique(x[endcap])
  plt.plot(r*np.ones(len(u)),u,'k.',markersize=1)

plt.show()
```
