```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import sympy as sym
import sympy.plotting.plot as symplot
from IPython.display import display, Math
import numpy as np
```

```python
x = 3
y = 5

plt.plot(x,y,'bs')

plt.axis('square')
plt.axis([-6,6,-6,6])
plt.grid()

plt.show()
```

```python
x,y = sym.symbols('x,y')
xpoints = range(-10,11)

expr = x**2 - 3*x

for i in xpoints:
  y = expr.subs(x,i)
  plt.plot(i,y,'o')

plt.xlabel('x')
plt.ylabel('f(x) = %s' %expr)
plt.show()
```

```python
x = [-5,5]
m = 2
c = 1

# Calculates without Numpy
# y = [0,0]
# for i in range(0,len(x)):
#   y[i] = m*x[i] + c

# Calculates with Numpy
y = m*np.array(x) + c

print(x,y)
plt.plot(x,y,label='y=%sx+%s' %(m,c))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')

plt.legend()

plt.show()
```

```python
def plot_line(x,m,c):
  y = m*np.array(x) + c
  plt.plot(x,y,label='y=%sx+%s' %(m,c))

plot_line([-5,5],.7,-2)
plot_line([-5,5],-1.25,.75)

plt.axis('square')
plt.grid()

plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```

# Rational Functions

```python
x = range(-3,4)
y = np.zeros(len(x))

for i in range(0,len(x)):
  y[i] = 2 - x[i]**2

plt.plot(x,y,'s-')
plt.show()
```

```python
x = np.linspace(-3,4,100)
y = 2 + np.sqrt(abs(x))

plt.plot(x,y,'m-')
plt.show()
```

```python
x = np.linspace(-4,4,100)

for power in range(-1,4):
  y = x**power
  plt.plot(x,y,linewidth=4,label='$y=x^{%s}$' %power)

plt.xlim([x[0],x[-1]])
plt.ylim([-20,20])

plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.show()
```

# Plotting with Sympy

```python
x = sym.symbols('x')

y = x**2

p = symplot(y,show=False)
p.xlim = [-10,10]
p[0].line_color = 'm'
p.title = 'Nice plot'

p.show()
```

```python
x,a = sym.symbols('x,a')

expr = a/x
subsa = 1

p = symplot(expr.subs(a,subsa),show=False)
p[0].label = '$y=%s$' %sym.latex(expr.subs(a,subsa))
p[0].line_color = 'r'

subsa = 2

p2 = symplot(expr.subs(a,subsa),show=False)
p.extend( p2 )
p[1].label = '$y=%s$' %sym.latex(expr.subs(a,subsa))

p.xlim = [-5,5]
p.ylim = [-5,5]
p.legend = True
p.show()
```

```python
x,a = sym.symbols('x,a')

y = a/(x**2 - a)

p = None

p = symplot(y.subs(a,1),show=False)
p[0].label = '$%s$' %sym.latex(y.subs(a,1))

for i in range(2,5):
  pnew = symplot(y.subs(a,i),show=False)
  p.extend( pnew )
  p[i-1].label = '$%s$' %sym.latex(y.subs(a,i))
  p[i-1].line_color = np.random.rand(3)
  
p.xlim = [-5,5]
p.ylim = [-10,10]

p.xlabel = 'x'

p.legend = True
p.show()
```

# Generating Images from Matrices

```python
A = [ [1,2],[1,4]]
plt.imshow(A)

plt.xticks([0,1])
plt.yticks([0,1])

plt.show()
```

```python
A = np.zeros((10,14))

shapeA = np.shape(A)
print(shapeA)

for i in range(0,shapeA[0]):
  for j in range(0,shapeA[1]):
    A[i,j] = 3*i - 4*j

plt.imshow(A)

plt.plot([0,3],[8,2],'r')

for i in range(0,shapeA[0]):
  for j in range(0,shapeA[1]):
    plt.text(j,i,int(A[i,j]),horizontalalignment='center',verticalalignment='center')

plt.set_cmap('twilight_shifted')

plt.show()
```

```python
board = np.zeros((10,10))

board_shape = np.shape(board)

for i in range(0,board_shape[0]):
  for j in range(0,board_shape[1]):
    board[i,j] = (-1)**(i+j)

plt.imshow(board)
plt.set_cmap('gray')
plt.tick_params(labelleft=False,labelbottom=False)

plt.show()
```

# Visualising Polygons

```python
from matplotlib.patches import Polygon

y = np.array([ [1,1],[2,3],[3,1] ])
p = Polygon(y,facecolor='m',alpha=.3)

y1 = np.array([ [2,2],[2.5,4],[3.5,1] ])
p1 = Polygon(y1,facecolor='b',alpha=.3,edgecolor='k')

fig,ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)

ax.set_ylim([0,5])
ax.set_xlim([0,4])

plt.show()
```

```python
# My Version:
x = np.linspace(-2,2,101)
fx = np.zeros(len(x))

for i in range(0,len(x)):
  fx[i] = -x[i]**2

points = []

for i in range(0,len(x)):
  points.append([x[i],fx[i]])

poly = Polygon(points,facecolor='g',alpha=.3,edgecolor='k',linewidth=1)
fig,ax = plt.subplots()
ax.add_patch(poly)

box = Polygon([ [-.5,-4], [-.5,-2.5], [.5,-2.5], [.5,-4] ],facecolor='k')
ax.add_patch(box)

ax.set_ylim([-4.25,0.25])
ax.set_xlim([-2.25,2.25])

plt.show()
```
