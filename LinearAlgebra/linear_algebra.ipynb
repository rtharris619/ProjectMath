```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
from mpl_toolkits.mplot3d import Axes3D
```

```python
row_vec = np.array([4,2])
col_vec = np.array([[-2],[3]])

display(Math(sym.latex(sym.sympify(row_vec))))
display(Math(sym.latex(sym.sympify(col_vec))))

plt.plot([0,row_vec[0]],[0,row_vec[1]],'r',label='row vector')
plt.plot([0,col_vec[0,0]],[0,col_vec[1,0]],'b',label='column vector')

plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(111,projection='3d')

vec1 = np.array([3,0,-4])
vec2 = np.array([-1,1,3])

ax.plot([0,vec1[0]],[0,vec1[1]],[0,vec1[2]],'b',linewidth=3)
ax.plot([0,vec2[0]],[0,vec2[1]],[0,vec2[2]],'r',linewidth=3)

ax.plot([-5,5],[0,0],[0,0],'k--',alpha=.25)
ax.plot([0,0],[-5,5],[0,0],'k--',alpha=.25)
ax.plot([0,0],[0,0],[-5,5],'k--',alpha=.25)

ax.set_xlim3d(-5,5)
ax.set_ylim3d(-5,5)
ax.set_zlim3d(-5,5)

# OR plt.axis([-5,5,-5,5,-5,5])

plt.show()
```

```python
v = np.array([.5,1])
s = [1,-.5,2,.5]

for si in s:
  sv = si*v
  plt.plot([0,sv[0]],[0,sv[1]],'o-',linewidth=3,label='$\\lambda=%g$'%si)

plt.axis('square')
plt.axis([-3,3,-3,3])
plt.grid()
plt.legend()
plt.show()
```

```python
v1 = np.array([-1,2])
v2 = np.array([1,1])

v3a = v1+v2
v3b = np.add(v1,v2)
v3c = np.zeros(2)

for i in range(2):
  v3c[i] = v1[i] + v2[i]

print(v3a)
print(v3b)
print(v3c)
```

```python
def get_latex(v):
  return sym.latex(sym.sympify(v))

v1 = np.array([-1,2])
v2 = np.array([1,1])
v3 = np.add(v1,v2)

plt.plot([0,v1[0]],[0,v1[1]],'b',linewidth=3,label='$v_1$')
plt.plot([v1[0],v1[0]+v2[0]],[v1[1],v1[1]+v2[1]],'orange',linewidth=3,label='$v_2$')
plt.plot([0,v3[0]],[0,v3[1]],'g',linewidth=3,label='$v_1+v_2$')

display(Math('%s+%s=%s' %(get_latex(v1),get_latex(v2),get_latex(v3))))

plt.axis('square')
plt.axis([-2,2,-2,4])
plt.legend()
plt.grid()
plt.show()
```
